```python
import pandas as pd

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")

gear_counts = df["# Gears"].value_counts() # conta a quantidade de vezes que cada valor aparece
gear_counts.plot(kind='bar')
```

```python
import seaborn as sns
sns.set_theme() 
gear_counts.plot(kind='bar')
```

```python
sns.histplot(df['CombMPG'], kde=True)
```

```python
carinfo = df[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']]
carinfo.head()
```

```python
sns.pairplot(carinfo, hue='Cylinders', height=2.5)
```

```python
sns.scatterplot(x='Eng Displ', y='CombMPG', data=df)
```

```python
sns.jointplot(x='Eng Displ', y='CombMPG', data=df, kind='scatter') 
# kind = tipo de grafo a ser plotado
```

```python
sns.lmplot(x='Eng Displ', y='CombMPG', data=df, hue='Cylinders', height=6)
# lmplot é um scatterplot com uma linha de regressão linear
```

```python
sns.set(rc={'figure.figsize':(15, 5)}) # Define o tamanho da figura
ax = sns.boxplot(x='Mfr Name', y='CombMPG', data=df, hue='Mfr Name')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sns.set(rc={'figure.figsize':(25, 12)}) # Alterando o tamanho da figura
ax = sns.swarmplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sns.set(rc={'figure.figsize':(15, 5)})
ax = sns.countplot(x='Mfr Name', data=df, hue='Mfr Name')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
sns.set(rc={'figure.figsize':(25, 5)})
#df2 = df.pivot_table(index='Mfr Name', columns='Cylinders', values='CombMPG', aggfunc='mean')
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean')
sns.heatmap(df2, annot=True, fmt=".1f") # fmt=".1f" é utilizado para formatar os valores com uma casa decimal
```

```python
gear_mpg = df[['# Gears', 'CombMPG']]
gear_mpg.head()
```

```python
sns.set(rc={'figure.figsize':(15, 5)})
sns.scatterplot(x='# Gears', y='CombMPG', data=gear_mpg)
```

```python
sns.lmplot(x='# Gears', y='CombMPG', data=gear_mpg, height=6)
```

```python
sns.jointplot(x='# Gears', y='CombMPG', data=gear_mpg, kind='scatter')
```

```python
sns.boxplot(x='# Gears', y='CombMPG', data=gear_mpg, hue='# Gears')
```

```python
sns.swarmplot(x='# Gears', y='CombMPG', data=gear_mpg)
```

```python
sns.set(rc={'figure.figsize':(15, 10)})
df3 = df.pivot_table(index='Eng Displ', columns='# Gears', values='CombMPG', aggfunc='mean')
sns.heatmap(df3, annot=True, fmt=".1f")
```

A partir dos gráficos é possível concluir que: as trasmissões mais populares são 8, 6 e 7. Os carros mais econômicos possuem 6 e 1 marchas. Não é possível assumir que com o aumento de marchas, o consumo aumenta, devido ao fato de ser implicado pelo tamanho do motor. É possível perceber também o aumento do consumo com o aumento de marchas em motores 2.0, por exemplo. O caso contrário acontece com motores 2.5.
