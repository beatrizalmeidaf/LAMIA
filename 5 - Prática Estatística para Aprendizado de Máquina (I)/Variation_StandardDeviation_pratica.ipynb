## Variância e Desvio Padrão

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(120)
incomes = np.random.normal(200.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std(), incomes.var()
```

```python
incomes = np.random.normal(200.0, 30.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes = np.random.normal(100.0, 150.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std(), incomes.var()
```
