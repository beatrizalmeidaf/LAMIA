## Percentis

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(150)
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
np.percentile(vals, 99)
```

```python
np.percentile(vals, 50)
```

```python
np.random.seed(150)
vals = np.random.normal(0, 5, 1000)

plt.hist(vals, 50)
plt.show()
```

```python
np.percentile(vals, 50)
```

```python
np.percentile(vals, 90)
```

## Momentos


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
# Momento de primeira ordem
np.mean(vals)
```

```python
# Momento de segunda ordem
np.var(vals)
```

```python
# Momento de terceira ordem
import scipy.stats as sp
sp.skew(vals)
# Indica que possui uma cauda mais acentuada a esquerda
```

```python
# Momento de quarta ordem
sp.kurtosis(vals)
```
