```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")

# Explorar as primeiras linhas do DataFrame
print("Primeiras linhas do dataset:")
print(df.head())

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando o estilo
sns.set_style("whitegrid")

# Gráfico de barras para o número de cilindros
plt.figure(figsize=(8, 5))
sns.countplot(x='Cylinders', data=df, palette='muted')
plt.title('Distribuição de Carros por Número de Cilindros')
plt.xlabel('Número de Cilindros')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de dispersão
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Eng Displ', y='CityMPG', hue='Cylinders', size='CombMPG', data=df, palette='cool')
plt.title('Deslocamento do Motor vs. City MPG')
plt.xlabel('Deslocamento do Motor (Litros)')
plt.ylabel('City MPG')
plt.legend(title='Cilindros')
plt.show()
```

```python
# Boxplot para CombMPG por Transmissão
plt.figure(figsize=(10, 6))
sns.boxplot(x='Transmission', y='CombMPG', data=df, palette='Set3')
plt.title('Eficiência Combinada por Tipo de Transmissão')
plt.xlabel('Transmissão')
plt.ylabel('CombMPG')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np

# Selecionando apenas colunas numéricas
num_cols = ['Eng Displ', 'Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG', '# Gears']

# Calculando correlações
corr = df[num_cols].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Mapa de Correlação')
plt.show()
```

```python
# Gráfico de barras para HwyMPG por Fabricante
plt.figure(figsize=(12, 6))
sns.barplot(x='Mfr Name', y='HwyMPG', data=df, palette='viridis', ci=None)
plt.title('Consumo em Estrada por Fabricante')
plt.xlabel('Fabricante')
plt.ylabel('HwyMPG')
plt.xticks(rotation=45)
plt.show()
```
